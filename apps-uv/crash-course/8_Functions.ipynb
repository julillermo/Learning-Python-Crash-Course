```python
# include a docstring to your function (code taken from ForrestKnight, Youtube)
# give intuitive variable names

# Use ALL_CAPS for constant variables
DEFAULT_LENGTH = 12
DEFAULT_WIDTH = 7

def calculate_area(length, width): # Well-named function parameters
    """
    Caclulate the area of a rectangle

    This function multiplies the length by the width to find the area of a rectangle.
    It's a straightforward mathematical calculation, but using descriptive variable names
    enhances readability and understanding.

    Args:
    length (int): The length of the rectangle.
    width (int): The width of the rectangle.

    Returns:
    int: The area of the rectangle
    """
    return length * width

# Using constants
area = calculate_area(DEFAULT_LENGTH, DEFAULT_WIDTH)
print(f"Area of rectangle with default sizes: {area}")
```

```python
# Positional Arguments
# This is when the order of the arguments should match the function definition

def describe_pet(animal_type, pet_name):
    """Display information about a pet."""
    
    print(f"\nI have a {animal_type}")
    print(f"My {animal_type}'s name is {pet_name.title()}")

describe_pet('hamster','harry')
# Here, animal_type corresponds to 'hamster' while
# pet_name corresponds to 'harry'

# order matters in positional arguments
```

```python
# Keyword Arguments
# when you already specify that argument variable a value belongs to

def describe_pet(animal_type, pet_name):
    """Display information about a pet."""
    print(f"\nI have a {animal_type}")
    print(f"My {animal_type}'s name is {pet_name.title()}")

# This works by explicity stating which value is assigned to which
# during the function call
describe_pet(pet_name='harry', animal_type='hamster')
```

```python
# Default Values

# function arguments can have default value
# If the user doesn't pass a value during function call, the default value is assumed

# however, default value arguments must be placed after regular positional arugments
#   this helps avoid confusion with the positional argument ordering

# normal posotional arguments are required for a functional call to work

def describe_pet(pet_name, animal_type='dog'):
    """Display information about a pet."""
    print(f"\nI have a {animal_type}")
    print(f"My {animal_type}'s name is {pet_name.title()}")

# Triggering just the minimum for the function to work
#   which is the positional argument
describe_pet('tulio')

# We don't have to use a keyword here, but it works either way
describe_pet(pet_name='harry')

# Even if we don't explicity state it, positional arguments will be followed
describe_pet('julius', 'weirdo')

# We can still override the argument as usual
describe_pet(pet_name='willie', animal_type='hamster')
```

```python
# Make optoinal arguments using Default Values

def get_formatted_name (first_name, last_name, middle_name=''):
    """Return a full name, neatly formatted."""

    # I think it's the really more of the if-else code
    #   that helps make the arguments optional
    if middle_name:
        full_name = f"{first_name} {middle_name} {last_name}"
    else:
        full_name = f"{first_name} {last_name}"
    return full_name.title()

musician = get_formatted_name('jimi', 'hendrix')
print(musician)

musician = get_formatted_name('john', 'hooker', 'lee')
print(musician)
```

```python
# Passing lists to functions

# The functions can directly modify the contents of
# a list passed as an argument

# Note that we already changed the argument names
# yet the code inside the function still altered the lists
def print_models(unprinted_designs_f, completed_models_f):
    while unprinted_designs_f:
        current_design = unprinted_designs_f.pop()
        print(f"Printing Model: {current_design}")
        completed_models_f.append(current_design)
    print()

def show_completed_models(completed_models):
    for completed_model in completed_models:
        print(f"Completed Model: {completed_model}")
    print()

# Note that the lists are OUTSIDE of the functions yet they are changed by it
# This is different from passing single variables that simply become copies and remain unaltered
unprinted_designs = ['phone case', 'robot pendant', 'dodecahedron']
completed_models = []

print("Before running the functions:")
print(f"unprinted designs: {unprinted_designs}")
print(f"completed models: {completed_models}")
print()

print_models(unprinted_designs, completed_models)
show_completed_models(completed_models)

print("Outcome after running the functions:")
print(f"unprinted designs: {unprinted_designs}")
print(f"completed models: {completed_models}")
```

```python
# Preventing a function from modifying a list

def print_models(unprinted_designs_f, completed_models_f):
    while unprinted_designs_f:
        current_design = unprinted_designs_f.pop()
        print(f"Printing Model: {current_design}")
        completed_models_f.append(current_design)
    print()

def show_completed_models(completed_models):
    for completed_model in completed_models:
        print(completed_model)
    print()

unprinted_designs = ['phone case', 'robot pendant', 'dodecahedron']
completed_models = []

print("Before running the functions:")
print(f"unprinted designs: {unprinted_designs}")
print(f"completed models: {completed_models}")
print()

# using the slice notation only sends a copy of the list to the function
# this makes sense because slices likely output copies without altering the original list
print_models(unprinted_designs[:], completed_models[:])
show_completed_models(completed_models)

print(f"unprinted designs: {unprinted_designs}")
print(f"completed models: {completed_models}")
print()

print("Outcome after running the functions:")
print(f"unprinted designs: {unprinted_designs}")
print(f"completed models: {completed_models}")


""" 
Even though you can preserve the contents of a list by passing a copy
of it to your functions, you should pass the original list to functions
unless you have a specific reason to pass a copy. It's more efficient
for a function to work with an existing list to avoid using the time and
memory needed to make a separate copy, especially when you're working
with larger lists.
"""
```

```python
# Passing an arbitrary number of arguments

# this notation allows the function to make a tuple of the entries
def make_pizza(*toppings):
    """ Print the list of toppings that have been requested. """

    print("\nMaking a pizza with the following toppings:")
    print(f"\n {type(toppings)}")

    # unpacking using a loop
    for topping in toppings:
        print(f"- {topping}")

make_pizza('pepperoni')
make_pizza('mushrooms', 'green peppers', 'extra cheese')

# Even when we pass a list or a dict, it still ends up being a tuple
make_pizza(['bread', 'sauch', 'air'])
make_pizza({'bread', 'sauch', 'air'})
```

```python
# Mixing Positional and Arbitrary Arguments

# A combination of the previous variations to the standard python function
# note that the arguments with variation are placed at the end

# This creates a TUPLE which can be iterated over
def make_pizza(size, *toppings):
    """ Summarize the pizza we are about to make. """
    print(f"\nMaking a {size}-inch pizza with the following toppings:")
    for topping in toppings:
        print(f"- {topping}")
    print(type(toppings))

make_pizza(16, 'pepperoni')
make_pizza(12, 'mushrooms', 'green peppers', 'extra cheese')
```

```python
# Using Arbitrary Keyword Arguments
# Functions that accept as many key-value pairs as the function call provides

# Python creates an empty DICTIONARY to be filled with the key-value pairs
def build_profile(first, last, **user_info):
    """ Build a dictionary containing everything we know about a user. """
    user_info['first_name'] = first
    user_info['last_name'] = last
    print(type(user_info))
    return user_info # python dictionary

# the function call sends out keywords that aren't pre-determined
user_profile = build_profile('albert', 'einstein', 
                            location='princeton',
                            field='physics')

print(user_profile)
```

```python
# Accessing an element in a list using negative index
ls = [0,1,2,3,4,5]
print(ls[-1])
# not all programming languages have this
```

```python
# Removing items from a list by Value
motorycycles = ['honda', 'yamaha', 'suzuki', 'ducati']
motorycycles.remove('yamaha')
print(motorycycles)
```

```python
# Organizing a List
cars = ['bmw', 'audi', 'toyota', 'subaru']
cars.sort()
print(cars)
```

```python
# Organizing a list temporarily
cars = ['bmw', 'audi', 'toyota', 'subaru']
print(sorted(cars))
```

```python
# Reversing the order of a list
cars = ['bmw', 'audi', 'toyota', 'subaru']
cars.sort()
cars.reverse()
print(cars)
```

```python
# Using int() for NUMERICAL input
age = input("How old are you?")
print(int(age))
```

```python
# Data validation (from the internet)
while True:
   try:
       user_input = input("Enter an integer: ")
       validated_input = int(user_input)
       print(validated_input)
       break
   except ValueError:
       print("That's not an integer! Try again.")
```

```python
# Using a continue in a loop
# use a 'continue' statement to returns to the beginning of a loop

current_number = 0
while current_number < 10:
    current_number += 1
    if current_number % 2 == 0:
        continue
    print(current_number)
```

```python
# Filling a Dictionary with User Input
responses = {}
polling_active = True

while polling_active:
    name = input("\nWhat is your name?")
    response = input("Which mountain wold you like to climb someday?")
    responses[name] = response

    repeat = input("Would you like to let another person respond? (yes/ no)")
    if repeat == 'no':
        polling_active = False
# note that the above code doesn't account for the way Python dictionaries
#   overwrite entries with the same key

print("\n--- Poll Resutls ---")
for name, response in responses.items():
    print(f"{name} would like to climb {response}")
```

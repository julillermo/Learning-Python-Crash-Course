```python
# Checking if a value exists within a list

requested_toppings = ['mushrooms', 'onions', 'pineapple']
print('mushrooms' in requested_toppings)
print('pepperoni' in requested_toppings)
```

```python
# Checking if a value IS NOT in a list

banned_users = ['andrew', 'carolina', 'david']
user = 'marie'

print(user not in banned_users) # True because it's not in the list
```

```python
# Checking That a List Is Not Empty

requested_toppings = []

if requested_toppings: # Since the string is empty, this line is equivalent to 'False'
    print("list is not empty")
else:
    print("Are you sure you want a plain pizza?")
```

```python
# Using Multiple Lists

available_toppings = ['mushrooms', 'olives', 'green peppers'
                    'pepperoni', 'pineapple', 'extra cheese']

requested_toppings = ['mushrooms', 'french fries', 'extra cheese']

for requested_topping in requested_toppings:
    if requested_topping in available_toppings:
        print(f"Adding {requested_topping}")
    else:
        print(f"Sorry, we don't have {requested_topping}")

print('\nFinished making your pizza!')
```

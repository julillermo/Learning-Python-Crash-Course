# Ignoring Files

Python automatically generates *.pyc* files from the *.py* files and stores then inside the *\_\_pycache\_\_* folder.

We can tel git to ignore these files through the **.gitignore**. To do this we create a *.gitignore* file with the following entry:<br>
&emsp;`__pycach__/` - referring to the *\_\_pycache\_\_/* folder and its contents

# Initializing a Resposity

Navigate to your desired folder through the terminal and send the following command, ***`git init`*** <br>

This command will initialize the use of **git** for the specified directory. This folder is now identified a repository. A <u>**repository**</u> is a set of files actively being tracke by git.

After initializing, an invisible file **.git** that Git uses to manage the repository.

# Making a Commit

Isse the command `git commit -m "message"`. <br>
* *the -m flag tells Git to record the message that follows in the project log*

Afterward, run the `git status` command again.

# Checking the Log

All commits are recording in the log of the project. Each commit is assigned a 40-character reference ID, information on who made the commit, when, and the recording message

# The Second Commit

When we run the `git status` command after making changes to any of our files, Git is able to pick up that the files have changed.

Run the `git commit -am "message."` command.
* The `-a` flag tells Git to add all modified files in the repo (files being monitored) to the current commit
* The `-m` flag tells Git to record the succeeding message.<br>

*As with terminal commands, you can separate or combine flags* 

# Reverting a change

This is for when we make a change to our code (say .py), saved the file, Git gets information that a file has changed, and we want to undo what we did.

We will 'undo' the changes without directly using the 'ctrl+z' or 'undo' features in our editor.

To do this, run the `git checkout .` command. Because of the `'.'`, this reverts everything to the latest commit.

Once we head back to the file we changed, we can see that the latest modifcations have been removed, and the original (latest commit) file is restored.

# Checking Out Previous Commits

*"You can check out any commit in your log, not just the most recent, by including the first six characters of the reference ID instead of a dot [ ' . ' ]"*

To view the reference IDs of the previous commits, simply use `git log` or `git log --pretty=oneline`

Checking out a previous commit exits you from the **master branch** into a ***detached HEAD*** state.

To get back to the master branch, send the `git checkout master` command

# To select 1 version and delete all other commits

on the master branch, run `git reset --hard <first-six-chars-of-reference-ID>`

Once we check the commits using `git log`, we notice that there is only one version left, the version that we selected

# Deleting the repository

You can delete a project's history by deleting the `.git` directory. The latest saved versions of the repository files will persist, but all the history of commits will be lost.

Deleting `.git` can be done manually through a file browser or command line

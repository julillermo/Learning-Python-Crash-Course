```python
# range is a datatype
r = range(1,6)
print(type(r))
print(len(r))
print()

# the range only reaches up to n-1 of the 2nd input
for i in range(0,10):
    print(i)
print()

# use np.linspace() if you want a list with a defined number of elements
import numpy as np
for i in np.linspace(1,10,19):
    print(i)
```

```python
# List Comprehensions

# making a list using programming logic
squares = [value**2 for value in range(1, 11)]
print(squares)

# Can also be achieved with lambda/anonymous functions
squares = [(lambda x: x**2)(n) for n in range(1, 11)]
print(squares)
```

```python
# Slicing a list
players = ['charles', 'martina', 'michael', 'florence', 'eli']
print(players[2::1]) # [start:end:step]
print(players[:4:1]) # [start:end:step]
print(players[1:4:1]) # [start:end:step]
```

```python
# Slicing a list
players = ['charles', 'martina', 'michael', 'florence', 'eli']
print(players[0:3])
print(players[1:4])

# I think the 1st arguement, the 'start' use 0 as the reference
#    while the 2nd argument, the 'end' uses 1 as the reference
# Alternatively, it's also possible to view the 2nd argument as
#   how many steps to the right of the chosen 'start' reference,
#   where '0' is the default if empty
```

```python
# looking through a slice
players = ['charles', 'martina', 'michael', 'florence', 'eli']
for player in players[:3]:
    print(player.title())
```

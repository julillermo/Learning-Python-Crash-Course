From my learning in JS, it would seem that Python dictionaries (aka hash or associative arrays) are alread considerd as "OJBECTS".
Python dicts can also apparently hold functions, similar to JS objects.

```python
# A simple dictionary
alien_0 = {
    'color': 'green',
    'points': 5
}

print(alien_0['color'])
print(alien_0['points'])
```

```python
# Adding New Key-Value Pairs
alien_0 = {
    'color': 'green',
    'points': 5
}

alien_0['x_position'] = 0
alien_0['y_position'] = 25
print(alien_0)

# to properly print this, I think we might need the JSON and CSV standard libraries
```

```python
# Modifying Values in a Dictionary

alien_0 = {'color': 'green'}
print(f"the alien is {alien_0['color']}")

alien_0['color'] = 'yellow'
print(f"The alien is now {alien_0['color']}")
```

```python
# More complex example for dictionaries
# This honestly comes close to what I understand as OOP

alien_0 = {
    'x_position': 0,
    'y_position': 25,
    'speed': 'medium'
}
print(f"Original position: {alien_0['x_position']}")

if alien_0['speed'] == 'slow':
    x_increment = 1
elif alien_0['speed'] == 'medium':
    x_increment = 2
else:
    x_increment = 3

alien_0['x_position'] = alien_0['x_position'] + x_increment

print(f"New position: {alien_0['x_position']}")
```

```python
# Removing Key-Value Pairs
alien_0 = {
    'color': 'green',
    'points': 5
}
print(alien_0)

del alien_0['points']
# FWIU, Python will automatically resize itself so there is no empty points.

print(alien_0)
```

```python
# A list as a value to a key-value pair
favorite_languages = {
    'jen': ['Python', 'C'],
    'edward': 'ruby',
    'phil': 'python',
}

favorite_languages
```

```python
# Using get() to Access Values

# in the get() function, if the first value doesn't exist,
# you can send the 2nd value instead

alien_0 = {'color': 'green', 'speed': 'slow'}

point_value = alien_0.get('points', 'No point value assigned.')
# The .get() approach acts like an if statement.
# If the first statement does not exist, output the second statement
print(point_value)
```

```python
# Looping through a dictionary

user_0 = {
    'username': 'efermi',
    'first': 'enrico',
    'last': 'fermi',
}

# The syntax below helps upack the key-value pairs of the dictionary
for key, value in user_0.items():
    print(f"\nKey: {key}")
    print(f"Value: {value}")

print(f"\n {user_0.items()}")
```

```python
# Looping through all the KEYS in a dictionary
favorite_languages = {
    'jen': 'Python',
    'sara': 'C',
    'edward': 'ruby',
    'phil': 'python',
}

# Using .keys() to extract just the keys()
for name in favorite_languages.keys():
    print(name.title())
```

```python
# Looping through all VALUES in a dictionary
favorite_languages = {
    'jen': 'Python',
    'sara': 'C',
    'edward': 'ruby',
    'phil': 'Python',
}

# use .values() to retrieve only the keys of the dictionary
print("The following languages have been mentioned:")
for language in set(favorite_languages.values()):
    print(language.title())
# the set() function removes duplicates (case-sensitive) and arranges the
    # list in order
```

```python
# A list of dictionaries
alien_0 = {'color': 'green', 'points': 5}
alien_1 = {'color': 'yellow', 'points': 10}
alien_2 = {'color': 'red', 'points': 15}

aliens = [alien_0, alien_1, alien_2]

for alien in aliens:
    print(alien)
```

```python
# A list in a dictionary
pizza = {
    'crust': 'thick',
    'toppings': ['mushrooms', 'extra cheese'],
}

print(f"You orderd a {pizza['crust']}-crust pizza with the following toppings:")
for topping in pizza['toppings']: # we can operate on the list as we usually do
    print(f"\t{topping}")
```

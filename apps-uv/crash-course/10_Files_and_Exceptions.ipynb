# Exceptions

```python
# Exceptions are handled with 'try-except' blocks
# When you use 'try-expect' blocks, your programs will continue running
#   even if things start to go wrong.
```

```python
# Handling the ZeroDivisionError Exception

print(5/0)
# In the error report below, ZeroDivisionError is an excpetion object
# These exceptions guide us on how to modify our program
```

```python
# Use try-except blocks

# We essentially tell Python what to do when there's a error.

try:
    print(5/0)
except ZeroDivisionError:
    print("You can't divide by zero!")
# note that we used the same exception object in our 'except' statement
#   as what was thrown in the above error

# ZeroDivisionError is actually a more specific error object of a "larger"
#   error object referred to as 'ArithmeticError'

# look at this diagram I found online for the choices of exception objects:
#   https://www.section.io/engineering-education/exceptions-and-error/exception_hierarchy.png

# You can also look at the Built-in Excpetions listed in the Python Documentation here:
#   https://docs.python.org/3/library/exceptions.html
```

```python
# the try-except block can actually work without specifying an error object
try:
    print(5/0)
except:
    print("You can't divide by zero!")
# from my testing, this only works for this specific example
# not specifying the exception object might result in unexpected results.
```

```python
# Using Excpetions to Prevent Crashes

# The code below doesn't yet include a way to deal with division by 0
print("Give me two numbers, and I'll divide them.")
print("Enter 'q' to quit.")

while True:
    first_number = input("\nFirst number: ")
    if first_number == 'q':
        break
    second_number = input("\nSecond number: ")
    if second_number == 'q':
        break
    answer = int(first_number) / int(second_number)
    print(answer) 
```

```python
# The Else Block

# updating the code to handle division by zero
while True:
    first_number = input("\nFirst number: ")
    if first_number == 'q':
        break
    second_number = input("\nSecond number: ")
    if second_number == 'q':
        break

    # Essentially this acts like an if-else statement specifically for errors
    try:
        answer = int(first_number) / int(second_number)
    except ZeroDivisionError:
        # After outputting the text, the while loop repeats
        #   after reaching the end of the block
        print("You can't divide by Zero")
    else:
        # when no exception is met, proceed with the else block
        print(answer)  
```

```python
# FileNoteFoundError corrected

filename = 'alice.txt'

try:
    with open(filename, encoding='utf-8') as f:
        contents = f.read()
except FileNotFoundError:
    print(f"Sorry, the file {filename} does not exist.")
```

```python
# Analyzing Text

# If you're working on a project that involves handling classic literray
#   text, Project Gutenberg host free texts

def count_words(filename):
    try:
        with open(filename, 'r') as temp:
            contents = temp.read()
    except FileNotFoundError:
        print(f"Sorry, the file {filename} does not exist")
    else:
        words_list = contents.split()
        wordcount = len(words_list)
        print(f"The file {filename} has about {wordcount} words.")

# filename = 'art_of_war.txt'
filename = 'communist_manifesto.txt'
count_words(filename)
```

```python
# Failing Silently

# Include a 'pass' statement inside the except block
# Doing so essentially skips over the input with an error
#   and repeats if its inside a loop

def count_words(filename):
    try:
        with open(filename, 'r') as temp:
            contents = temp.read()
    except FileNotFoundError:
        pass
    else:
        words_list = contents.split()
        wordcount = len(words_list)
        print(f"The file {filename} has about {wordcount} words.")

files = ['art_of_war.txt', 'communist_manifesto.txt', 'alice.txt', 'iliad.txt']
for file in files:
    count_words(file)
```

# Storing Data

```python
# A common way to store user data is through the use of a JSON file
```

```python
# json.dump() and json.load()
import json

numbers = [2, 3, 5, 7, 11, 13]

filename = 'numbers.json'
with open(filename, 'w') as temp:
    json.dump(numbers, temp) # this is a list, however
```

```python
import json

filename = 'numbers.json'
with open(filename) as temp:
    numbers_load = json.load(temp)

print(numbers_load)
```

```python
# Directly and Reading User-Generated Data
import json

username = input("What is your name? ")

filename = 'username.json'
with open(filename, 'w') as temp:
    json.dump(username, temp)
    print(f"We'll remember you when you come back, {username}!")
```

```python
import json

filename = 'username.json'

with open(filename) as temp:
    username_load = json.load(temp)
    print(f"Welcome back, {username}!")
```

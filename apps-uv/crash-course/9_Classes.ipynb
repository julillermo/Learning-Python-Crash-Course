```python
class Dog: # Capitalize names of classes
    """A simple attempt to model a dog."""

    def __init__(self, name='Julius', age=69):
        """Initialize name and age attributes."""
        # These are attributes
        self.name = name
        self.age = age

    # class method functions can be built the same way as normal Py functions
    # but, always include 'self' as an argument even if it won't be used
    def sit(self):
        """Simulate a dog sitting in reponse to a command."""
        print(f"{self.name} is now sitting.")

    def roll_over(self):
        """Simulate rolling over in response to a command."""
        print(f"{self.name} rolled over!")
```

```python
""" The __init__(self, ...) Method """
# functions that are part of a class are called METHODS
# the 'self' parameter is required in the method definition.
# the 'self' param must come first in the arguments.
# the 'self' param gives access to attributes and methonds in the class

# notice that 'self' is repeatedly used in the succeeding functions even
#   when they aren't used

# variables that are accessible through instances are callsed ATTRIBUTES
```

```python
# Making an Instance from a Class

def output(dogo):
    print(f"\nMy dog's name is {dogo.name}")
    print(f"My dog is {dogo.age} years old.")

my_dog = Dog() # made an instance of the Dog class
output(my_dog)
my_dog.sit()
my_dog.roll_over()

my_dog2 = Dog('Tulio', 1_000) # an instance of the Dog class
output(my_dog2)
my_dog2.sit()
my_dog2.roll_over()

# Take note that attributes don't have parentheses 
# while methonds/functions do (have parentheses)
```

```python
# Inheretance in OOP

# Parent Class
class Car:
    """ A simpel attempt to represent a Car. """

    def __init__(self, make, model, year):
        """ Initialize attributes to describe a car. """
        self.make = make
        self.model = model
        self.year = year

        # attributes can be set without being part of the function call
        # I guess this is somewaht equivalent to placing it at the function definition 
        self.odometer_reading = 0
        self.gas_level = 0.2

    def get_descriptive_name(self):
        """ Return a neatly formatted descriptive name. """
        long_name = f"{self.year} {self.make} {self.model}"
        return long_name.title()

    def read_odometer(self):
        """ Print a statement showing the car's mileage. """
        print(f"This car has {self.odometer_reading} miles on it.")

    def update_odometer(self, mileage):
        """ 
        Set the odometer reading to the given value. 
        Reject the change if it attempst to roll the odometer
        """
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back the odometer!")

    def increment_odometer(self, miles):
        """ Add the given amount to the odometer reading. """
        self.odometer_reading += miles

    def fill_gas_tank():
        """ Model filling up a car's gas until full """
        self.gas_level = 1.0


# Child Class of Car
class ElectricCar(Car): # include name of parent class in parenthesis
    """ Represent aspects of a car, specific to electic vehicles."""

    def __init__(self, make, model, year): # same function definition as parent class
        """
        Initialize attributes of the parent class. 
        Then initialize attributes specific to an electric car.
        """
        # the super() function lets you call functions form the parent class
        # all the parent functions are preloaded using super()
        super().__init__(make, model, year)

        self.battery_size = 75

    # you can start adding child class specific functions here
    def describe_battery(self):
        """ Print a statement describing the battery size. """
        print(f"This car has a {self.battery_size}-kWh battery")

    # to override a parent class method, just use the same method name in the child
    def fill_gas_tank(self):
        """ Electric cars don't have gas tanks. """
        print("This car doesn't need a gas tank!")

my_tesla = ElectricCar('tesla', 'model s', 2019)
print(my_tesla.get_descriptive_name()) # function not explicitly part of the child

# Directly changing attribute value
my_tesla.odometer_reading = 23
my_tesla.read_odometer()

# Change attribute value using a method
my_tesla.update_odometer(96)
my_tesla.read_odometer()
my_tesla.update_odometer(1)

my_tesla.increment_odometer(4)
my_tesla.read_odometer()

my_tesla.describe_battery()

my_tesla.fill_gas_tank()
```

```python
# Class methods can be made to become a separate class

# Parent Class
class Car:
    """ A simpel attempt to represent a Car. """

    def __init__(self, make, model, year):
        """ Initialize attributes to describe a car. """
        self.make = make
        self.model = model
        self.year = year

        # attributes can be set without being part of the function call
        # I guess this is somewaht equivalent to placing it at the function definition 
        self.odometer_reading = 0
        self.gas_level = 0.2

    def get_descriptive_name(self):
        """ Return a neatly formatted descriptive name. """
        long_name = f"{self.year} {self.make} {self.model}"
        return long_name.title()

    def read_odometer(self):
        """ Print a statement showing the car's mileage. """
        print(f"This car has {self.odometer_reading} miles on it.")

    def update_odometer(self, mileage):
        """ 
        Set the odometer reading to the given value. 
        Reject the change if it attempst to roll the odometer
        """
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back the odometer!")

    def increment_odometer(self, miles):
        """ Add the given amount to the odometer reading. """
        self.odometer_reading += miles

    def fill_gas_tank():
        """ Model filling up a car's gas until full """
        self.gas_level = 1.0

# A Battery() object whose methods we can load in a another class
class Battery:
    """ A simple attempt to model a battery for an electric car. """

    def __init__(self, battery_size=75):
        """ Initialize the battery's attributes. """
        self.battery_size = battery_size

    def describe_battery(self):
        """ Print a statement desribing the battery size. """
        print(f"This car has a {self.battery_size}-kWh battery")

# Child Class of Car
class ElectricCar(Car): # include name of parent class in parenthesis
    """ Represent aspects of a car, specific to electic vehicles."""

    def __init__(self, make, model, year): # same function definition as parent class
        """
        Initialize attributes of the parent class. 
        Then initialize attributes specific to an electric car.
        """
        # the super() function lets you call functions form the parent class
        super().__init__(make, model, year)
        self.battery = Battery() # referred to the class Battery()
        # recall that classes are capitalized
        # From what I understand, we're using another class as a folder
        # for different functions. This is more organized in terms of
        # viewing things as "objects"

        # NOTE on this:
        # - I think, instead of loading the Battery() class as an attribute,
        #   We can just make the child class inherit from multiple classes
        # - This approach would be disorganized, however.

    # to override a parent class method, just use the same method name in the child
    def fill_gas_tank(self):
        """ Electric cars don't have gas tanks. """
        print("This car doesn't need a gas tank!")

my_tesla = ElectricCar('tesla', 'model s', 2019)
print(my_tesla.get_descriptive_name()) # function not explicitly part of the child

my_tesla.battery.describe_battery()
# The ".battery" is an attribute that calls the "Battery" class
# The "Battery" class contaings the "describe_battery() function"
```
